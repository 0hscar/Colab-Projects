```python
import pandas as pd
from linearmodels.panel import PanelOLS, RandomEffects, compare
import statsmodels.api as sm
```

```python
data = pd.read_csv('Data_kandidatavhandling.csv', delimiter=';')

data['Total Debt / Total Assets '] = data['Total Debt / Total Assets '].str.replace(',','.').str.rstrip('%').astype('float') / 100
data['ROCE'] = data['ROCE'].str.replace(',','.').str.rstrip('%').astype('float') / 100
data['Current ratio'] = data['Current ratio'].str.replace(',','.').str.rstrip('%').astype('float') / 100
data['log (assets)'] = data['log (assets)'].str.replace(',','.').astype('float')
data['Short Term Debt / Total Assets'] = data['Short Term Debt / Total Assets'].str.replace(',','.').str.rstrip('%').astype('float') / 100
data['Long Term Debt / Total Assets'] = data['Long Term Debt / Total Assets'].str.replace(',','.').str.rstrip('%').astype('float') / 100

data.set_index(['Company code', 'Year'], inplace=True)


dependent_vars = ['Total Debt / Total Assets ', 'Short Term Debt / Total Assets', 'Long Term Debt / Total Assets']
independent_vars = ['ROCE', 'Current ratio', 'log (assets)']

for dep_var in dependent_vars:
  model = PanelOLS(data[dep_var], sm.add_constant(data[independent_vars]), entity_effects=True, time_effects=False, other_effects=data['COVID (Dummy)'])
  # model = PanelOLS(data[dep_var], data[independent_vars])

  results = model.fit()
  print(results)
data
```
