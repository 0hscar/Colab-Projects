# 4.C top 10 filmer med mer än 40 ratings

```python
topTen = ratingData[ratingData.itemID.isin(ratingData.itemID.value_counts().index[ratingData.itemID.value_counts().gt(40)])].groupby("itemID")["rating"].mean().sort_values(ascending=False).head(10)
topTen

```

# 5. A Det är ingen skillnad vilken "column" man väljer att  sätta som tredje så visar den endå amount, därför renamea ja den bara (Skulle kunna sätta userID och den visar endå amounten)

```python
jobPerGender = userData.groupby(["Job", "Gender"]).count().reset_index().rename(columns={"zipcode": "Amount"})
jobPerGender[["Job", "Gender", "Amount"]]
```

10 populäraste yrken för män

```python
jobPerGender[["Job", "Gender", "Amount"]].loc[jobPerGender["Gender"] == "M"].sort_values(by="Amount", ascending=False).head(10).reset_index()
```

10 populäraste yrken för kvinnor

```python
jobPerGender[["Job", "Gender", "Amount"]].loc[jobPerGender["Gender"] == "F"].sort_values(by="Amount", ascending=False).head(10).reset_index()
```
