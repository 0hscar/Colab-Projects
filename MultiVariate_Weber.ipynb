Multi variate case (As2) Weber


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x1 = np.array([1.5, 1.5, 1.5, 1.5, 1.5, 2.5, 5, 2.5, 2.5, 5, 5]).reshape((-1,1))
x2 = np.array([1.5, 1.5, 1.5,  2.5, 5, 1.5, 1.5, 2.5, 2.5, 5, 5]).reshape((-1,1))
x3 = np.array([1.5, 2.5, 5, 1.5, 1.5, 1.5, 1.5, 1.5, 2.5, 2.5, 5]).reshape((-1,1))
y = np.array([30.70, 23.75, 6.25, 50.75, 144.5, 50.75, 100.75, 70.75, 63.75, 207.5, 190]).reshape((-1, 1))

poly = PolynomialFeatures(degree=2, include_bias=False)
lin_reg = LinearRegression()

```

```python
x_poly = poly.fit_transform(np.column_stack((x1, x2, x3)))

lin_reg.fit(x_poly, y)
lin_reg.score(x_poly, y)
```

```python
x_test = poly.transform([[5, 5, 1.5]])
y_pred = lin_reg.predict(x_test)
y_pred
```
