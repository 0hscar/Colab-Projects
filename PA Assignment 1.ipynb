```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('Book1.csv', sep=';', header=None, skiprows=1, names=['date', 'price'])
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%y')

# Create plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df['date'], df['price'])

# Format plot
ax.set_xlabel('Date')
ax.set_ylabel('Share Price (â‚¬)')
ax.set_title('Historical Share Prices')
plt.xticks(rotation=45)
plt.grid()

# Display plot
plt.show()
```
