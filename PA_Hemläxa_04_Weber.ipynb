```python
import matplotlib.pyplot as plt


def initialize_towers(num_disks):
    towers = [[], [], []]
    for i in range(num_disks, 0, -1):
        towers[0].append(i)
    return towers

# Five colors if you want to run it with five disks, the choice is made later at "num_disks"
def visualize_towers(towers):
    plt.clf()
    colors = ['blue', 'green', 'red', 'purple', 'orange']
    max_height = max(len(tower) for tower in towers)
    fig, ax = plt.subplots(figsize=(6, 4))
    for i, tower in enumerate(towers):
        for j, disk in enumerate(tower):
            plt.Circle((i + 1, max_height - len(tower) + j + 1), disk * 0.1, color=colors[disk - 1], ec='black')
            plt.gca().add_patch(plt.Circle((i + 1, max_height - len(tower) + j + 1), disk * 0.1, color=colors[disk - 1], ec='black'))
    plt.xlim(0, 4)
    plt.ylim(0, max_height + 1)
    plt.xticks([])
    plt.yticks([])
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Tower of Hanoi')
    plt.show()
    plt.pause(0.5) # Little pause so you can see somewhat whats happening

# Solving function
def tower_of_hanoi(n, source, target, auxiliary):
    if n > 0:
        # Move n-1 disks from source to auxiliary, using target as auxiliary
        tower_of_hanoi(n - 1, source, auxiliary, target)

        # Move the nth disk from source to target
        target.append(source.pop())

        # Current state
        visualize_towers(towers)

        # Move the n-1 disks from auxiliary to target using source as auxiliary, check function parameters to see whats going on
        tower_of_hanoi(n - 1, auxiliary, target, source)

# Initialize towers with disks
num_disks = 3
towers = initialize_towers(num_disks)

# Starting state for the towers
visualize_towers(towers)

tower_of_hanoi(num_disks, towers[0], towers[2], towers[1])

```
