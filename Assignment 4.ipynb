```python
import spacy

paragraph = "Albert Einstein was a German-born physicist who developed the theory of relativity. He is considered one of the most influential scientists of the 20th century. Einstein was born in Ulm, Germany in 1879 and died in 1955 in the United States. He was awarded the Nobel Prize in Physics in 1921."
question = "When was Albert Einstein born?"

nlp = spacy.load("en_core_web_sm")

doc = nlp(paragraph)
question_doc = nlp(question)

# Find the answer to the question in the paragraph
answer = None
for ent in doc.ents:
    if ent.label_ == "PERSON" and ent.text in question_doc.text:
        for ent2 in doc.ents:
            if ent2.label_ == "DATE" and "born" in ent2.sent.text.lower() and ent.start <= ent2.start:
                # If the person is mentioned in the question and the DATE is mentioned after the person, extract the DATE
                # Swap "DATE" to "GPE" and change the question to "Where was ..." and it answers that question
                answer = ent2.text
                break
        if answer:
            break

if answer:
    print("The answer is:", answer)
else:
    print("Sorry, I don't know the answer.")
```
