```python
import nltk

from nltk.chunk import ChunkParserI, conlltags2tree
from nltk.chunk.util import tree2conlltags

swedish_words = [
    'björn', 'är', 'en', 'stor', 'däggdjur',
    'brukade', 'leva', 'i', 'skogar',
    'nu', 'finns', 'mest', 'i', 'zoo',
    'äter', 'hö', 'och', 'bär',
    'har', 'tjock', 'päls', 'för', 'att', 'skydda', 'mot', 'köld',
    'är', 'en', 'solskensdjur', 'och', 'älskar', 'att', 'sova',
    'björnen', 'är', 'mycket', 'stark', 'och', 'snabb',
    'kan', 'springa', 'fort', 'och', 'klättra', 'i', 'träd',
    'är', 'en', 'god', 'simmare', 'och', 'dykare',
    'älskar', 'att', 'leka', 'med', 'sin', 'ungar'
]


swedish_tagged_words = nltk.pos_tag(swedish_words)
print('Swedish Tagged Words: ', swedish_tagged_words)

class SwedishChunkParser(ChunkParserI):
    def __init__(self, chunked_sents):
        self.chunked_sents = chunked_sents

   
    def parse(self, tagged_sent):
        chunks = []
        chunk = []
        for word, tag in tagged_sent:
            if tag == 'NNP':
                chunk.append((word, tag))
            elif tag == 'VBP' and chunk:
                chunk.append((word, tag))
            elif tag == 'NN' and chunk:
                chunk.append((word, tag))
                chunks.append(chunk)
                chunk = []
        return chunks


swedish_chunk_parser = SwedishChunkParser(swedish_tagged_words)


chunks = swedish_chunk_parser.parse(swedish_tagged_words)
print('Swedish Chunks: ', chunks)
```
